## Module and Data Imports

```python
import os
import tarfile
import urllib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DOWNLOAD_ROOT = 'https://raw.githubusercontent.com/ageron/handson-ml2/master/'
HOUSING_PATH = os.path.join('datasets', 'housing')
HOUSING_URL = DOWNLOAD_ROOT + 'datasets/housing/housing.tgz'
```

```python
def fetch_housing_data(housing_url = HOUSING_URL, housing_path = HOUSING_PATH):
    os.makedirs(housing_path, exist_ok= True)
    tgz_path = os.path.join(housing_path, 'housing.tgz')
    
    # get url to the directory
    urllib.request.urlretrieve(url = housing_url, filename = tgz_path)
    
    # open the zip file and extract it
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
def load_housing_data(housing_path = HOUSING_PATH):
    fetch_housing_data()
    csv_path = os.path.join(housing_path, 'housing.csv')
    return pd.read_csv(csv_path)
```

```python
housing = load_housing_data()
```

## Splitting the Dataset to Train/Test

Here, we use Stratified-sampling method to have same distribution in the train/test sets similar to original dataset based on median_income field. 

```python
# Since, median_income is a continuous field, we first make a new field called, income_cat, which is a numerical ordinal.

housing['income_cat'] = pd.cut(x = housing['median_income'], bins = [0, 1.5, 3, 4.5, 6, np.inf], labels = [1,2,3,4,5])
```

```python
housing.head()
```

```python
housing['income_cat'].hist()
```

```python
# Let's use stratified sampling.

from sklearn.model_selection import StratifiedShuffleSplit

# Provides train/test indices to split data in train/test sets.

# This cross-validation object is a merge of StratifiedKFold and
# ShuffleSplit, which returns stratified randomized folds. The folds
# are made by preserving the percentage of samples for each class.
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)

for train_index, test_index in split.split(X = housing, y = housing['income_cat']):
    # print(train_index, test_index)
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
# Let's see how the distribution in the stratified dataset

strat_train_set['income_cat'].value_counts()/strat_train_set.shape[0]
```

```python
# Let's see how the distribution in the overall dataset.
housing['income_cat'].value_counts()/housing.shape[0]
```

```python
# how would the data distribution, if we were to use just random sampling

from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 42)

train_set['income_cat'].value_counts()/train_set.shape[0]

# Obviously data is not much different because it's drawing from the same dataset. It would follow similar distribution,
# but stratified split is more akin to original distribution and random split would result slightly skewed data distribution.
```

```python
# Let's now remove the income_cat from the train/test sets because it is not important now.

for dataset in [strat_train_set, strat_test_set]:
    dataset.drop('income_cat', axis = 1, inplace = True)
```

```python
strat_train_set.head()
```

## Data Cleaning

### Separate Dependent & Independent Variables

```python
def separate_dependent_feature(dataset, dependent_feature):
    independent_features = dataset.drop(labels = dependent_feature, axis = 1)
    dept_feature = dataset[dependent_feature]
    
    return independent_features, dept_feature
```

```python
train_x, train_y = separate_dependent_feature(dataset = strat_train_set, dependent_feature='median_house_value')
```

### Handling Missing Values

```python
# We saw that there are some null values in the dataset. 

housing.isnull().sum()
```

```python
# We have few options to handle missing values.
# 1. Drop those instances/observations/records altogether.
# 2. Drop that feature altogether. - This is probably a consideration when significantly high % missing
# 3. Fill missing values with another value, either median, mean, most common, etc.

# We will handle missing values for both numerical and categorical features using scikit-learn SimpleImputer.

from sklearn.impute import SimpleImputer

numerical_imputer = SimpleImputer(strategy = 'median')
categorical_imputer = SimpleImputer(strategy= 'most_frequent')
```

```python
numerical_features = train_x.select_dtypes(include = 'number')
print('Numerical Features:', numerical_features.columns, '\n')

categorical_features = train_x.select_dtypes(exclude = 'number')
print('Categorical Features:', categorical_features.columns)
```

```python
# Let's use this numerical imputer to fill the null values in the numerical features. 
imputed_num_set = numerical_imputer.fit_transform(strat_train_set[numerical_features.columns])
imputed_num_set = pd.DataFrame(imputed_num_set, columns=numerical_features.columns)
imputed_num_set.isnull().sum()
```

```python
# Let's use this categorical imputer to fill the null values in the categorical features. In here, we don't have missing values.
# But, we don't know what is the case for future inputs. So, we should better handle that scenario.

imputed_cat_set = categorical_imputer.fit_transform(strat_train_set[categorical_features.columns])
imputed_cat_set = pd.DataFrame(imputed_cat_set, columns=categorical_features.columns)
imputed_cat_set.isnull().sum()
```

### Handling Categorical Variables

```python
# Two options for categorical - ordinal encoding or one-hot encoding.

# eg. from sklearn.preprocessing import OrdinalEncoder
# However, with the ordinal encoder, numerical encoding implies to the model that there is a certain relation in the data.
# So, it is better to use such encoding, when there is actually such relationship in the data.

# Otherwise, it's recommended to use one-hot encoding.
# For this we can use, pandas get_dummies() function or sklearn's OneHotEncoder.

from sklearn.preprocessing import OneHotEncoder

categorical_encoder = OneHotEncoder() # we may need to drop one encoded column to avoid dummy variable trap.

housing_cat_1hot = categorical_encoder.fit_transform(train_x[['ocean_proximity']])
```

### Custom Transformer - Feature Engineering (Or Creating New Features from Existing)

```python
# Let's define a class that implement the transformer, for the purpose of creating few additional features.
# We can use this new transformer in sklearn pipeline.

from sklearn.base import BaseEstimator, TransformerMixin

# define the indices for feature columns
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_vs_rooms = True):
        self.add_bedrooms_vs_rooms = add_bedrooms_vs_rooms
    
    def fit(self, X, y = None):
        return self # nothing else to do, we would have to do somethings, if we setting any parameters for future reference
        # like for instance, median or etc, which we might need in future transform. But, in this case, this transformer 
        # is not something that looks at overall dataset, it will just manipulate instances individually in the dataset. 
    
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix]/X[:, households_ix]
        population_per_household = X[:, population_ix]/X[:, households_ix]
        
        if self.add_bedrooms_vs_rooms:
            bedrooms_vs_rooms = X[:, bedrooms_ix]/X[:, rooms_ix]
            
            # return column-wise concatenated numpy array
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_vs_rooms]

        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

```python
attr_adder = CombinedAttributesAdder(add_bedrooms_vs_rooms=True)
housing_extra_attributes = attr_adder.transform(strat_train_set.values)
```

```python
pd.DataFrame(housing_extra_attributes)
```

### Feature Scaling

This is important for linear models and neural networks. Not so much of a problem for decision tree algorithms.

Two popular methods, min-max scaler and standard scaler. 

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
standard_scaler = StandardScaler()

standard_scaler.fit_transform(train_x[numerical_features.columns].values)
```

## Transformation Pipelines

```python
from sklearn.pipeline import Pipeline

numerical_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')), 
                              ('attribute_adder', CombinedAttributesAdder()),
                               ('standard_scaler', StandardScaler())])

numerical_pipeline.fit_transform(train_x[numerical_features.columns])
```

```python
# Instead of having one numerical and categorical pipelines, we can combine those two.

from sklearn.compose import ColumnTransformer

numerical_columns = list(numerical_features.columns)
categorical_columns = list(categorical_features.columns)
```

```python
full_pipeline = ColumnTransformer([('numerical', numerical_pipeline, numerical_columns),
                                   ('categorical', OneHotEncoder(), categorical_columns)
                                  ])

# we will now send training X data through the Pipeline
train_x_transformed = full_pipeline.fit_transform(train_x)
```

```python
train_x_transformed
```

## Model Training

### Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()

# fitting the model to data
linear_regressor.fit(train_x_transformed, train_y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Let's see how this perform with the training dataset
y_pred_train = linear_regressor.predict(train_x_transformed)

rmse = np.sqrt(mean_squared_error(train_y, y_pred_train))
print('Root MSE of Linear Regression:', rmse)
```

```python
# So prediction error is $68,628.
```

### Decision Tree Regressor 

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(train_x_transformed, train_y)
```

```python
# Let's try the performance with training dataset 
y_pred_train = dt_regressor.predict(train_x_transformed)

rmse = np.sqrt(mean_squared_error(train_y, y_pred_train))
print('Root MSE of Decision Tree Regressor:', rmse)
```

```python
# This shows a badly overfit situation for the decision tree regression model. 
```

## Cross-Validation for Better Model Evaluation

```python
def print_scores(scores, model_name):
    print(f'Statistics for {model_name} model based on cross-validation:\n')
    print('Scores:', scores, '\n')
    print('Mean:', scores.mean(), '\n')
    print('Standard Deviation:', scores.std())
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator=linear_regressor, 
                         X = train_x_transformed, 
                         y=train_y, 
                         cv=10, 
                         scoring='neg_mean_squared_error')

lin_reg_rmse_scores = np.sqrt(-scores)

print_scores(lin_reg_rmse_scores, 'Linear Regression')
```

```python
scores = cross_val_score(estimator=dt_regressor, 
                         X = train_x_transformed, 
                         y=train_y, 
                         cv=10, 
                         scoring='neg_mean_squared_error')

dt_reg_rmse_scores = np.sqrt(-scores)

print_scores(dt_reg_rmse_scores, 'Decision Tree Regression')
```

## Hyper-parameter Tuning with Grid Search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
param_grid = [{'n_estimators': [3, 10, 30], 'max_features': [2,4,6,8]},              
              {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]}             
             ]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(estimator=forest_reg, 
                           param_grid=param_grid, cv=5, 
                           scoring = 'neg_mean_squared_error',
                           return_train_score=True
                          )

grid_search.fit(train_x_transformed, train_y)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
# if you want to see the evaluation scores for each

cvres = grid_search.cv_results_
```

```python
# Let's print these results for each grid point

for mean_test_score, params, mean_train_score in zip(cvres['mean_test_score'], cvres['params'], cvres['mean_train_score']):
    print(params, ' Test MSE: ', -mean_test_score, 'Train MSE: ', -mean_train_score)

```

```python
# same thing with a more readable Pandas dataframe
pd.DataFrame({'Parameters': cvres['params'], 
              'Mean Training Score': -cvres['mean_train_score'],
              'Mean Test Score': -cvres['mean_test_score']
             })
```

## Hyper-parameter Tuning with Random Search

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = [{'n_estimators': [10, 30, 50,100,200], 'max_features': [6,8, 10, 15]},              
              {'bootstrap': [False], 'n_estimators': [3, 10,50,100], 'max_features': [2, 3, 4, 10]}             
             ]
```

```python
# so here we have total parameter space of 36, but we trying out 20 out of those 36, chosen randomly.

random_search = RandomizedSearchCV(estimator=forest_reg, param_distributions=param_grid, 
                                   n_iter=20, scoring = 'neg_mean_squared_error', cv=5)

random_search.fit(train_x_transformed, train_y)
```

```python
random_search.best_params_
```
